# Data Collection — Nguồn dữ liệu và nội dung

Mục tiêu: trả lời câu hỏi 'Dữ liệu của nhóm lấy từ đâu và gồm những gì?'. Dưới đây là tóm tắt các nguồn chính repository đang sử dụng (tham khảo `data_sources/`).

## Tóm tắt ngắn (trả lời trực tiếp)

- Dữ liệu của nhóm lấy từ nhiều corpus public liên quan tới email spam/ham: SpamAssassin Public Corpus (Apache), SetFit/enron_spam (Hugging Face), locuoco 300k (Hugging Face).
- Ngoài ra có các nguồn tham khảo như TREC Spam Track và CMU Enron Dataset (dùng làm tham khảo/ mở rộng nếu cần).
- Nội dung: các file CSV đã xử lý (`data/processed/`) chứa bảng các email với các cột thường gặp như `id`, `subject`, `body`/`text`, và cột nhãn (`spam`/`ham` hoặc 0/1). Thư mục `data/raw/` chứa bản thô (một file = một email) cần parse khi tái tạo pipeline.

## Nguồn dữ liệu (chi tiết từ `data_sources/data_links.csv`)

1) SpamAssassin Public Corpus — https://spamassassin.apache.org/old/publiccorpus/
   - Loại: Official Apache SpamAssassin archive.
   - Ghi chú: chứa các thư mục `easy_ham`, `hard_ham`, `spam` được dùng trong `src/data_loader.py` để sinh dữ liệu huấn luyện.

2) SetFit/enron_spam — https://huggingface.co/datasets/SetFit/enron_spam
   - Loại: Community-hosted trên Hugging Face.
   - Ghi chú: tập CSV khoảng ~33.7k email có nhãn; dùng để bổ sung dữ liệu đã gán nhãn.

3) locuoco/the-biggest-spam-ham-phish-email-dataset-300000 — https://huggingface.co/datasets/locuoco/the-biggest-spam-ham-phish-email-dataset-300000
   - Loại: Community-hosted lớn trên Hugging Face.
   - Ghi chú: dataset lớn (spam/ham/phish); trong project mapping nhãn phish/spam → `spam` để phù hợp bài toán binary.

## Nguồn tham khảo thêm (không bắt buộc)

- TREC Spam Track — https://trec.nist.gov/data/spam.html (official): dùng làm tham chiếu cho corpus học thuật/benchmark nếu cần.
- CMU Enron Email Dataset — https://www.cs.cmu.edu/~enron/ (official): chủ yếu email công ty/ham; có thể dùng làm nguồn bổ sung sau khi gán nhãn phù hợp.

(Các nguồn này được ghi chép đầy đủ trong `data_sources/OFFICIAL_SOURCE_REVIEW.md` với đánh giá về tính chính thống và cách sử dụng.)

## Nội dung dữ liệu trong repository

- `data/processed/` — CSV đã gom và (có thể) tiền xử lý: mỗi hàng là một email; cột thường gặp: `id`, `subject`, `body`/`text`, `label` (ví dụ `spam`/`ham` hoặc 0/1). File tiêu biểu: `combined_labeled.csv`.
- `data/processed/*_balanced.csv` — phiên bản đã cân bằng lớp cho thử nghiệm so sánh.
- `data/raw/` — email thô (một file = một email) theo nguồn; cần parse để trích text/subject và gán nhãn khi chưa có.
- `data_sources/data_links.csv` và `data_sources/OFFICIAL_SOURCE_REVIEW.md` — mô tả nguồn gốc, link và đánh giá độ chính thống; dùng để minh bạch nguồn dữ liệu.

## Kết luận (ngắn) — trả lời câu hỏi

- Dữ liệu của nhóm lấy từ các corpus public: SpamAssassin (Apache), các dataset public trên Hugging Face (SetFit/enron_spam, locuoco 300k), và tham khảo thêm TREC & Enron khi cần.
- Dữ liệu gồm: email (subject/body/text) và nhãn spam/ham; lưu trữ ở dạng CSV đã xử lý trong `data/processed/` và bản thô trong `data/raw/` nếu cần tái xử lý.
