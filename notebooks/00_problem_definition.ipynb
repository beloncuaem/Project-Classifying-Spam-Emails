# Problem Definition

Tài liệu này mô tả vấn đề mà nhóm sẽ giải trong project phân loại email spam vs ham.

## 1. Bối cảnh (Background)

- Nhiều email chứa nội dung không mong muốn (spam) gây phiền nhiễu người dùng và rủi ro bảo mật.
- Mục tiêu: xây dựng mô hình phân loại email thành `spam` hoặc `ham` để lọc tự động và cải thiện trải nghiệm người dùng.

## 2. Mục tiêu (Objective)

- Phát triển pipeline tiền xử lý, đặc trưng hóa, và huấn luyện mô hình phân loại email.
- So sánh hiệu năng giữa các mô hình (ví dụ: Naive Bayes, Logistic Regression, SVM).
- Cung cấp metric rõ ràng để đánh giá (precision/recall/F1, ROC-AUC) cho lớp `spam`.

## 3. Phạm vi (Scope)

- Dữ liệu sử dụng: tập dữ liệu đã được lưu trong thư mục `data/processed/` (ví dụ `combined_labeled.csv`).
- Không bao gồm: phát hiện phishing nâng cao, phân tích liên kết hoặc xử lý attachment đa phương tiện (ngoại trừ body/text).
- Kết quả cần: model file, báo cáo metric, và notebook trình bày EDA + kết quả.

## 4. Dữ liệu (Data)

- Các file nguồn chính nằm ở `data/processed/` và `data/raw/`.
- Cột quan tâm: text của email (body/subject), nhãn (spam/ham).
- Kiểm tra cân bằng lớp, missing values, và chất lượng văn bản (encoding, noise).

## 5. Tiêu chí thành công (Success metrics)

- Primary: F1-score cho lớp `spam` >= 0.85 (tùy dataset, đặt mục tiêu ban đầu).
- Secondary: Precision cao để giảm false positives; ROC-AUC cho tổng quan model.
- Triển khai: mô hình có thể lưu/ nạp bằng `joblib`/`pickle` và chạy inference trên email mới.

## 6. Ràng buộc & Giả định (Constraints & Assumptions)

- Dữ liệu text tiếng Anh chủ yếu; không tối ưu cho nhiều ngôn ngữ khác.
- Không sử dụng mô hình lớn (LLM) do giới hạn tài nguyên — ưu tiên mô hình nhẹ/ classical.
- Thời gian huấn luyện và inference phải phù hợp với máy cá nhân (no GPU required).

## 7. Deliverables

- Notebook: `notebooks/00_problem_definition.ipynb` (văn bản vấn đề).
- Notebook: EDA, feature engineering, và training (sẽ tạo tiếp).
- Trained model files trong `models/` và báo cáo metric trong `reports/`.
